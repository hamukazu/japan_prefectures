```python
import pandas as pd
```

```python
df = pd.read_csv("a.txt", header=None, sep="\t")
```

```python
df2 = df[0].str.split(expand=True).rename(columns={0:"id", 1:"prefecture_jp"})
```

```python
def f(pref):
    if pref == "北海道":
        return "道"
    elif pref == "東京":
        return "都"
    elif pref in ["大阪", "京都"]:
        return "府"
    else:
        return "県"
df2["suffix_jp"] = df2.prefecture_jp.apply(f)
```

```python
df2 = df2.join(df[1]).rename(columns={1: "prefecture_en"})
```

```python
df2["region_jp"] = None
df2["region_en"] = None
```

```python
rcols = ["region_jp", "region_en"]
df2.loc[df2.id=="01", rcols] = "北海道","Hokkaido"
df2.loc[(df2.id>="02") & (df2.id<="07"), rcols] = "東北地方", "Tohoku"
df2.loc[((df2.id>="08") & (df2.id<="14")) | df2.id.isin(["19", "20"]), rcols] = "関東地方", "Kanto"
df2.loc[(df2.id>="15") & (df2.id<="18"), rcols] = "北陸地方", "Hokuriku"
df2.loc[(df2.id>="21") & (df2.id<="24"), rcols] = "中部地方", "Chubu"
df2.loc[(df2.id>="25") & (df2.id<="30"), rcols] = "近畿地方", "Kinki"
df2.loc[(df2.id>="31") & (df2.id<="35"), rcols] = "中国地方", "Chugoku"
df2.loc[(df2.id>="36") & (df2.id<="39"), rcols] = "四国地方", "Shikoku"
df2.loc[(df2.id>="40") & (df2.id<="46"), rcols] = "九州地方", "Kyushu"
df2.loc[df2.id=="47", rcols] = "沖縄県", "Kyushu"
```

```python
df2 = df2[["id", "region_jp", "region_en", "prefecture_jp", "suffix_jp", "prefecture_en"]]
```
